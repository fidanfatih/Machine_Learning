**Hopkins Test**
* Zaman ve maliyet verimlilii iin, lk olarak Veride kmeleme eilimi var mi yok mu belirlemek iin ``Hopkins Testi`` yaplr.
* Null Hypothesis(Ho) ve Alternative Hypothesis(Ha) temeline dayaniyor.
* Null Hypothesis(Ho): Uniform dagilim var, anlamli kmeleme yok.
* Alternative Hypothesis(Ha):  Veri, ratsgele veri noktalarindan oluur. Yani Kumeleme vardir.
* [0,1] araliginda bir score verir. score, 0a yaklatka veri uniform degil,yani clusteringe meyilli
* 1e yaklatka uniform yapi var, 0.5 gecmedikce kmelenebilir olarak ifade edilir. pratikte 0.3 sinir alinir.

# **Elbow Method**
* aciklanan varyans(Sum of squared distances) ve kme sayisi(k) arasindaki ilikiye dayanarak cozum gelitiren bir yntem, 
* yani aciklanan varyans, kme sayisinin fonksiyonu olarak izilir, grafikte dirsein kirildigi nokta, k nin optimal degeridir
* minimum k ile minimum hatanin alindigi optimum noktayi baz alacagiz. Keskin dususun en son bittigi yeri alacagiz. 

```python
ssd = []
K = range(1,10)
for k in K:
    kmeans = KMeans(n_clusters = k).fit((df))
    ssd.append(kmeans.inertia_)
    
```

```python
plt.figure(figsize=(15,6))
plt.plot(K, ssd, "bx-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error")
plt.title("Elbow Method");
```
